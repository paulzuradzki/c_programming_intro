```python
def read_file(f):
    with open(f, 'r') as file_obj:
        for index, line in enumerate(file_obj.readlines()):
            print(str(index+1)+'\t', line, end='')
```

```python
for f in l:
    if f.endswith('.c'):
        print('='*10,f, '='*10, sep=' ')
        read_file(f)
        print('\n')
```
